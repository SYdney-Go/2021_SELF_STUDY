## 회귀분석

```python
# 최소제곱
import numpy as np

x = [2, 4, 6, 8]
y = [81, 93, 91, 97]

# x와 y의 평균
mx = np.mean(x)
my = np.mean(y)
print('x의 평균값: ', mx)
print('y의 평균값: ', my)

# a = x - x의 평균 합의 제곱
divisor = sum([(mx - i)**2 for i in x])
```

```python
# 기울기 공식의 분자
def top(x, mx, y, my):
    d = 0
    for i in range(len(x)):
        d += ((x[i] - mx) * (y[i] - my))
    return d
#dividend = top(x, mx, y, my)
dividend = sum((x[i] - mx) * (y[i] - my) for i in range(len(x)))
print('분모: ', divisor)
print('분자: ', dividend)

# 기울기와 y 절편 구하기
a = dividend / divisor
b = my - (mx * a)

print('기울기 a =', a)
print('y 절편 b =', b)
```

#### 최소제곱법 실습

```python
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [0.2, 0.3, 0.5, 0.6, 0.9, 0.95, 1.1, 1.5]

mx = np.mean(x)
my = np.mean(y)
print('x의 평균값: ', mx)
print('y의 평균값: ', my)

divisor = sum([(mx - i)**2 for i in x])
dividend = sum((x[i] - mx) * (y[i] - my) for i in range(len(x)))
print('분모: ', divisor)
print('분자: ', dividend)


a = dividend / divisor
b = my - (mx * a)
print('기울기 a =', a)
print('y 절편 b =', b)

def size(week):
    cal = a * week + b
    if cal < 30:
        return round(cal,3)
    else:
        return 30

# 예측 문제    
print('15주차 예상크기: ', size(15))
print('22주차 예상크기: ',size(22))
print('77주차 예상크기: ',size(77))
print('200주차 예상크기: ',size(200))
# 5주차 예상크기
print('5주차 예상크기와 실크기의 차 :',round((size(5) - 0.9), 3))
```
