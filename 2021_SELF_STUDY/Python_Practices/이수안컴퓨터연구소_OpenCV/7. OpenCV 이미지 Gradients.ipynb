```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

from PIL import Image
from io import BytesIO
import requests
```

# Image Gradients
* 영상의 edge 및 방향 찾기

### Sobel
* Gaussian smoothing과 미분을 이용
* noise가 있는 이미지에 적용
* x, y축을 미분하여 경계값 계산

```python
url = 'https://cdn.pixabay.com/photo/2020/02/20/19/48/travel-4865665_960_720.jpg'
response = requests.get(url)
image = Image.open(BytesIO(response.content))
image = np.asarray(image)
print(image.shape)
plt.imshow(image)
```

```python
image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(image, cmap='gray')
```

```python
sobelx = cv2.Sobel(image, cv2.CV_8U, 1, 0, ksize=3)
plt.imshow(sobelx, cmap = 'gray')
```

```python
 sobely = cv2.Sobel(image, cv2.CV_8U, 0, 1, ksize=3)
 plt.imshow(sobely, cmap='gray')
```

```python
sobelx2 = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5)
plt.imshow(sobelx2, cmap='gray')
```

```python
sobely2 = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)
plt.imshow(sobely2, cmap='gray')
```

### Laplacian 함수
* 이미지의 가로와 세로에 대한 gradient를 2차 미분한 값
* Sobel filter에 미분의 정도가 더해진 것과 비슷..

```python
laplacian = cv2.Laplacian(image, cv2.CV_8U)
plt.imshow(laplacian, cmap='gray')
```

```python
laplacian2 = cv2.Laplacian(image, cv2.CV_64F)
plt.imshow(laplacian2, cmap='gray')
```

### Canny Edge Detection
* Noise Reduction
* Edge Gradient Detection
* Non-maximum Suprression
* Hystersis Thresholding

```python
canny = cv2.Canny(image, 30, 70)
plt.imshow(canny, cmap='gray')
```

```python
canny = cv2.Canny(image, 200, 300)
plt.imshow(canny, cmap='gray')
```
